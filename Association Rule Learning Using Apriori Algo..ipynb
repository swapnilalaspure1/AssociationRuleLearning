```python
# Structure the data in the form of list

# transaction[0 --> 9689]

transactions = []

for i in range(0, len(tv)):
    
  # Create a list for the current row, converting each value to a string
    transaction = [str(value) for value in tv.iloc[i]]
    
    
    # Append the transaction list to the 'transactions' list
    transactions.append(transaction)
    
transactions
```

```python
# Library import
import apyori
print('Apyori Version Check :-' ,apyori.__version__)
```

```python
# Creating the rules using apriori
# I want at leats 3 people should watch a movie from the total entries, 
# Min support --: 3*7/9690
# Confidence  --: 0.2 = 20%

rules = apriori( transactions,
                min_support = 0.0021,
                min_confidence = 0.2,  
                min_lift = 3,
                max_length = 2)

result = list(rules)  
```

```python
### Creating a function to exract every information in the form of data frame
def inspect(data):
    LHS = [ tuple( result[2] [0] [0] )[0] for result in data ]
    
    RHS = [ tuple( result[2] [0] [1] )[0] for result in data ]
    
    Support_Value = [ result[1] for result in data ]
    
    Confidence_Value = [ result[2] [0] [2] for result in data ]
    
    Lift_Value = [ result[2] [0] [3] for result in data ]
    
    return list(zip( LHS, RHS, Support_Value, Confidence_Value, Lift_Value ))
```

```python
# Assuming 'result' contains your association rule mining results
df = pd.DataFrame(inspect(result), columns=["LHS", "RHS", "Support_values", "Confidence_values", "Lift_values"])
df
```
